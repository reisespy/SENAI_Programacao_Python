Escreva um programa que pergunte a velocidade do
carro de um usuário. Caso ultrapasse 80 km/h, exiba uma
mensagem dizendo que o usuário foi multado. Nesse
caso, exiba o valor da multa, cobrando R$ 5 por km acima
de 80 km/h.

```python
# -*- coding: utf-8 -*-
"""
Código desenvolvido por Ariel Reises
No curso de Programação Python do Senai Alfried Krupp - Campo Limpo Paulista
01 de agosto de 2023
"""

# Apresentação inicial
print("Olá, este programa perguntará a velocidade do veículo do usuário")
print("E caso exceda a velocidade permitida, o multará em R$ 5,00 por km excedente")
print("")

# Criação das variáveis
velocidade = int(input("Insira a velocidade do veículo (em km/h): "))
if velocidade > 80:
  valor_multa= (velocidade-80) * 5
  print("")
  print("Usuário foi multado!")
  print("")
  print(f"O condutor ultrapassou em {velocidade-80}km/h a velocidade máxima permitida da via.")
  print(f"O valor da multa é de R${valor_multa:.2f}")
else:
  print("")
  print("Usuário não foi multado.")
  print("Obrigado por respeitar os limites de velocidade.")

# Assinatura - Ariel Reises
print("")
print("Obrigado por utilizar o sistema")
print("Copyright © 2023 - reises.py")
```

Escreva um programa para aprovar o empréstimo
bancário para a compra de uma casa. O programa deve
perguntar o valor da casa a comprar, o salário e a
quantidade de anos a pagar. O valor da prestação mensal
não pode ser superior a 30% do salário. Calcule o valor da
prestação como sendo o valor da casa a comprar dividido
pelo número de meses a pagar.

```python
# -*- coding: utf-8 -*-
"""
Código desenvolvido por Ariel Reises
No curso de Programação Python do Senai Alfried Krupp - Campo Limpo Paulista
02 de agosto de 2023
"""
import math

# Apresentação inicial
print("Olá, este programa calculará o valor de prestação do emprestimo bancário para compra de um imóvel.")
print("")

# Criação de variáveis
valor_casa = float(input("Insira o valor do imóvel desejado: R$ "))
salario = float(input("Insira o valor do salário mensal: R$ "))
anos = int(input("Insira a quantidades de anos desejados para o parcelamento: "))
meses = (anos / 12)
valor_prestacao = valor_casa / (anos * 12)

# Condições e apresentação
if valor_prestacao > salario * 0.30:
    print("")
    print("Emprestimo negado!")
    print("Infelizmente o valor da prestação excede o limite de desconto salarial.")
    print(f"Para sua faixa salarial, o valor máximo de parcela disponível é de R$ {(salario * 0.30):.2f}")
elif valor_prestacao < salario * 0.30:
    print("")
    print("Emprestimo aprovado!")
    print("Parabéns, você obteve liberação para a compra do imóvel.")
    print(f"O valor da prestação será de R$ {math.ceil(valor_prestacao):.2f}, e o pagamento será realizado em {(anos*12)} meses.")

# Assinatura - Ariel Reises
print("")
print("Obrigado por utilizar o sistema")
print("Copyright © 2023 - reises.py")
```

Escreva um programa que calcule o preço a pagar pelo
fornecimento de energia elétrica. Pergunte a quantidade
de kWh consumida e o tipo de instalação: R para
residências, I para indústrias e C para comércios. Calcule o
preço a pagar, de acordo com a tabela a seguir.

```python
# -*- coding: utf-8 -*-
"""
Código desenvolvido por Ariel Reises
No curso de Programação Python do Senai Alfried Krupp - Campo Limpo Paulista
02 de agosto de 2023
"""

# Apresentação inicial
print("Olá, este programa calculará o valor à pagar pelo fornecimento de energia elétrica.")
print("")
print("Utilize: R para Residências, C para Comércios e I para Indústrias.")
print("")

# Criação das variáveis de entrada
instalacao = input("Insira o tipo de instalação (R, C ou I): ")
kwh = float(input("Insira a quantidade consumida (em kWh): "))

# Cálculo do valor a pagar
if instalacao == "R" or "r":
  if kwh <= 500:
    preco = 0.40
  elif kwh > 500:
    preco = 0.65
elif instalacao == "C" or "c":
  if kwh <= 2500:
    preco = 0.55
  elif kwh <= 5000:
    preco = 0.60
elif instalacao == "I" or "i":
  if kwh <= 10000:
    preco = 0.55
  elif kwh <= 15000:
    preco = 0.60
else:
  print("Insira um tipo de instalação válida!")

# Impressão do valor à pagar
print("")
print(f"O valor à pagar é: R$ {(kwh * preco):.2f}.")

# Mensagens de alerta
if instalacao == "R" or "r" and kwh > 1000:
  print("")
  print(f"Atenção: Revise sua faixa de consumo mensal.")
elif instalacao == "C" or "c" and kwh > 5000:
  print("")
  print(f"Atenção: Revise sua faixa de consumo mensal.")
elif instalacao == "I" or "i" and kwh > 15000:
  print("")
  print(f"Atenção: Revise sua faixa de consumo mensal.")

# Assinatura - Ariel Reises
print("")
print("Obrigado por utilizar o sistema")
print("Copyright © 2023 - reises.py")
```
