1. Crie um programa para efetuar a leitura de um número inteiro e apresentar o resultado do quadrado deste número.

Passo à passo:


1.   Criar a variável de input do número desejado
2.   Criar a variável com a formula ao quadrado (numero * numero) ou número ***2
3.   Apresentar o resultado.



```python
#Leitura de número inteiro e apresentação do resultado do quadrado do número.

#Criando as variáveis:
numero = float(input("Insira o número desejado: "))
quadrado = (numero * numero)
    #também pode ser apresentada por: quadrado = numero**2

#Apresentando o resultado:
print("")
print(f"O número escolhido foi: {numero}, e esse número elevado ao quadrado é: {quadrado}")

#Assinatura
print("")
print("")
print("Obrigado por utilizar nosso sistema - © reises.py - 2023")
```

3. Crie um programa que leia um número inteiro e imprimir seu sucessor e seu antecessor.

Passo à passo:
1.   Criar a variável de input do número desejado (usando int)
2.   Criar as variáveis de antecessor (-1) e sucessor (+1)
3.   Apresentar o resultado



```python
#Antecessor e sucessor.
print("Olá! Escolha um número inteiro, e apresentarei o antecessor e o sucessor deste número.")
print("")

#Criando as variáveis:
numero = int(input("Insira o número desejado: "))
ant = numero-1
suc = numero+1

#Apresentando o resultado:
print("")
print(f"O número escolhido foi: {numero}")
print("")
print(f"Seu antecessor é {ant}")
print(f"Seu sucessor é {suc}")

#Assinatura
print("")
print("")
print("Obrigado por utilizar nosso sistema - © reises.py - 2023")
```

4. Crie um programa para entrar com a base a altura de um retângulo e imprimir respectivamente o perímetro e a área correspondente.

```python
#Perímetro e Área de um retangulo.
print("Olá! Informe a base e altura de um retângulo, e te informarei qual o perímetro e a área correspondente.")
print("")

#Criando as variáveis:
base = int(input("Insira o valor da base: "))
altura = int(input("Insira o valor da altura: "))
perimetro = (base+altura)*2
area = base*altura

#Apresentando o resultado:
print("")
print(f"A base do seu retangulo é {base} e a altura é {altura}.")
print("")
print(f"Seu perímetro é {perimetro}")
print(f"Sua área é {area}")

#Assinatura
print("")
print("")
print("Obrigado por utilizar nosso sistema - © reises.py - 2023")
```

5. Crie um programa que dados o valor, a taxa e o tempo, efetuar o cálculo do valor de uma prestação em atraso, utilizando a fórmula: prestação = valor + (valor * (taxa/100) * tempo)

```python
#Calculo de valor de prestação em atraso:

print("Olá! Te ajudarei no calculo de uma prestação em atraso.")
print("")

#Criando as variáveis:
valor = float(input("Insira o valor devido (em reais): R$ "))
taxa = float(input("Insira a taxa de juros a.m (em %): "))
tempo = int(input("Insira o tempo de atraso (em dias): "))
prestacao = valor + (valor * (taxa/100) * (tempo/30))

#Apresentando o resultado:
print("")
print(f"O valor inicial devido é: R$ {valor}")
print(f"Com a taxa de juros referente à {tempo} dias, o valor corrigido é: R$ {prestacao:.2f}")

#Assinatura
print("")
print("")
print("Obrigado por utilizar nosso sistema - © reises.py - 2023")
```
