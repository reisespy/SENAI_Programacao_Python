# Exemplos - Parte 02

```python
#Exemplo 8 - Tabuada simples

n = int(input("Tabuada do: "))
x = 0
while x <= 10:
  print(f"{n} x {x} = {n*x}")
  x = x + 1
```

```python
#Exemplo 9 - Soma de 10 números, com acumulador.

n = 1
soma = 0
while n <= 10:
  x = int(input("Digite o %d número: " % n))
  soma = soma + x
  n = n + 1
print ("Soma: %d" % soma)
```

```python
#Exemplo 10 - Cálculo da média de 5 números, com acumulador.

x = 1
soma = 0
while x <= 5:
  n = int(input("%d Digite o número: " % x))
  soma = soma + n
  x = x + 1
print ("Média: %5.2f" % (soma/5))
```

```python
#Exemplo 11 - Interrompendo a repetição, com break.

s = 0
while True:
  v = int(input("Digite um número a somar ou 0 para sair: "))
  if v == 0:
    break
  s = s + v
print (s)
```

```python
#Exemplo 12 - Tabuada aninhada.

tabuada = 1
while tabuada <= 10:
  n = 0
  print("Tabuada %d " % tabuada)
  while n <= 10:
    print("%d * %d = %d" % (tabuada, n, tabuada * n))
    n = n + 1
  tabuada = tabuada + 1
```

```python
#Exemplo 13 - Uso da função range(fim).

# Programa para imprimir 10 valores (de 0 a 9).
for v in range(10):
  print(v)
```

```python
#Exemplo 14 - Uso da função range(início, fim), com intervalos.

for v in range(5, 10):
  print(v)
```

```python
#Exemplo 15 - Uso da função range(início, fim, incremento), com saltos.

# Programa para imprimir os números 0, 2, 4, 6 e 8.
for v in range(0, 10, 2):
  print(v)
```

# 1) Entrar com números e imprimir o triplo de cada número. O programa acaba quando entrar o número - 999.

```python
numero = int(input("Insira o número inicial: "))
final = 999

while numero <= final:
  print(f"{numero}")
  numero = numero * 3
```

# 2) Entrar com números enquanto forem positivos e imprimir quantos números foram digitados.

```python
print("Para encerrar a inserção de números, digite 0.")
print("")
contador = 0
numero = int(input("Digite um número: "))
while numero != 0:
  if numero > 0:
    contador += 1
  numero = int(input("Digite o próximo número: "))
print("")
print(f"Foram digitados {contador} números.")
```

# 3) Entrar com vários números positivos e imprimir a média dos números digitados.

```python
quantidade = int(input("Digite a quantidade de números à serem somados: "))
contagem = 0
soma = 0

while True:
  if contagem < quantidade:
    valor = float(input("Digite o número: "))
    soma+= valor
  else:
    break
  contagem+=1

print(f"A média dos numeros informados é: {(soma/quantidade):.2f}")
```

# 4) Ler vários números e informar quantos números entre 100 e 200 foram digitados. Quando o valor 0 (zero) for lido, o programa deve cessar sua execução.

```python
print("Para encerrar a inserção de números, digite 0."); print("")
contador = 0; numero = int(input("Digite o primeiro número: "))
while numero != 0:
  if numero >= 100 and numero <= 200:
    contador += 1
  numero = int(input("Digite o próximo número: "))
print(""); print(f"Foram digitados {contador} números entre 100 e 200.")
```

# 5) Entrar com sexo de várias pessoas e imprimir quantas pessoas são do sexo masculino (considerar m ou M).
