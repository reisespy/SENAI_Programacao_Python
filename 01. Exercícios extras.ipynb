Escreva um código que calcule a hipotenusa de um triângulo retângulo, cujos catetos são a = 4 e b = 3.

```python
# -*- coding: UTF-8 -*-
"""
Código desenvolvido por Ariel Reises
No curso de Programação Python do Senai Alfried Krupp - Campo Limpo Paulista
01 de agosto de 2023
"""

#Apresentação inicial
print("Olá, calcularemos a hipotenusa de um triangulo retangulo.")
print("")
print("Informe abaixo as medidas conhecidas solicitadas.")

#Criação das variáveis
a_ca = 3                                          #float(input("Qual a medida do Cateto Adjacente? "))
b_co = 4                                          #float(input("Qual a medida do Cateto Oposto? "))
c_hip = ((a_ca * a_ca) + (b_co * b_co)) ** 0.5
                                                  #c_hip = (((a_ca**2) + (b_co**2))**0.5)

'''
O teorema de Pitágoras determina que o quadrado da medida da hipotenusa (C²)
é igual à soma dos quadrados das medidas dos catetos (A²+B²).
Portanto, a fórmula do teorema de Pitágoras é C²=A²+B²
'''

#Apresentação do resultado
print("")
print(f"a hipotenusa é: {c_hip:.2f}")

#Assinatura - Ariel Reises
print("")
print("Obrigado por utilizar o sistema")
print("Copyright © 2023 - reises.py")
```

Crie um programa e declare uma constante PI (use 4 casas após a vírgula). Dados o raio e a altura, calcular e apresentar o valor do volume de uma lata de óleo, utilizando a fórmula:
volume = PI * r 2 * altura

```python
# -*- coding: UTF-8 -*-
"""
Código desenvolvido por Ariel Reises
No curso de Programação Python do Senai Alfried Krupp - Campo Limpo Paulista
01 de agosto de 2023
"""

#Importação de biblioteca
import math

#Apresentação inicial
print("Olá, este programa calculará o valor do volume de uma lata de óleo a partir do seu raio e altura.")
print("")
print("Insira abaixo os valores solicitados")

#Criação das variáveis
pi = 3.14159265358979323846
raio = float(input("Insira o valor do raio (em cm): "))
altura = float(input("Insira o valor da altura (em cm) "))
area = (pi * (raio**2))
volume = (pi * (raio**2) * altura)
mililitros = volume
litros = mililitros / 1000

#Apresentação do resultado
print("")
print(f"A área de base da lata é de {math.floor(area):.2f} cm.")
print(f"O volume total da lata é de {math.floor(volume):.2f} cm³.")
print(f"Isso significa que caberá aproximadamente {math.floor(mililitros):.0f} ml na lata.") #
print(f"Isso em litros é igual à: {litros:.2f} L.")

#Assinatura - Ariel Reises
print("")
print("Obrigado por utilizar o sistema")
print("Copyright © 2023 - reises.py")
```

Calculo de hipotenusa utilizando a biblioteca math

```python
# -*- coding: UTF-8 -*-
"""
Código desenvolvido por Ariel Reises
No curso de Programação Python do Senai Alfried Krupp - Campo Limpo Paulista
01 de agosto de 2023
"""

#Apresentação inicial
print("Olá, calcularemos a hipotenusa de um triangulo retangulo com medidas dos catetos 3 e 4.")
print("")

#Importando a biblioteca math
import math

#Criação das variáveis
A = 3 #Cateto A
B = 4 #Cateto B
C = math.sqrt((A**2) + (B**2))

'''
O teorema de Pitágoras determina que o quadrado da medida da hipotenusa (C²)
é igual à soma dos quadrados das medidas dos catetos (A²+B²).
Portanto, a fórmula do teorema de Pitágoras é C²=A²+B²
'''

#Apresentação do resultado
print("")
print(f"a hipotenusa é: {C:.0f}")

#Assinatura - Ariel Reises
print("")
print("Obrigado por utilizar o sistema")
print("Copyright © 2023 - reises.py")
```
