9) Escreva um programa para calcular a redução do
tempo de vida de um fumante. Pergunte a quantidade de
cigarros fumados por dia e quantos anos ele já fumou.
Considere que um fumante perde 10 minutos de vida a
cada cigarro, calcule quantos dias de vida um fumante
perderá. Exiba o total em dias.

```python
# -*- coding: UTF-8 -*-
"""
Código desenvolvido por Ariel Reises
No curso de Programação Python do Senai Alfried Krupp - Campo Limpo Paulista
01 de agosto de 2023
"""

#Apresentação inicial
print("Olá, calcularemos a média de redução de vida de um fumante, baseado em quantos cigarros são fumados por dia, e quanto tempo ele fuma.")
print("")

#Criação das variáveis
cigarros_por_dia = int(input("Informe a quantidade de cigarros fumados por dia: "))
anos_fumando = int(input("Informe a quantidade de anos fumando: "))

#Criação da função para calcular a redução do tempo de vida
def calcular_reducao_tempo_vida(cigarros_por_dia, anos_fumando):  # Calcula a redução do tempo de vida de um fumante.
    total_cigarros = cigarros_por_dia * 365 * anos_fumando        # Argumento:
    tempo_perdido_minutos = total_cigarros * 10                   #     cigarros_por_dia (int): Quantidade de cigarros fumados por dia.
    reducao_em_dias = (tempo_perdido_minutos / (24 * 60))         #     anos_fumando (int): Quantidade de anos fumando.
    return reducao_em_dias                                                              # Retorna:
                                                                  #     float: Redução do tempo de vida em dias.

#Apresentação do resultado
if cigarros_por_dia == 0 and anos_fumando == 0:                   # Comparando se é igual à 0
  print(" ")
  print("Você não reduziu sua jornada de vida.")
  print(" ")
  print("Parabéns por não manter esse hábito! Cigarro faz mal à saúde.")
else:
  print(" ")
  print("Estimamos que você reduziu sua jornada em cerca de %5.2f dia(s)." % calcular_reducao_tempo_vida(cigarros_por_dia, anos_fumando))
  print(" ")
  print("O Ministério da Saúde adverte: Fumar causa câncer de pulmão.")

#Assinatura - Ariel Reises
print(" ")
print("")
print("Obrigado por utilizar o sistema")
print("Copyright © 2023 - reises.py")
```

bancário para a compra de uma casa. O programa deve
