**Criando uma Calculadora com Python**

```python
#Nome da Calculadora
print("Calculadora de adição ou subtração de dois produtos")

#Variáveis (utilizando o int para números)
num_1 = int(input("Insira o primeiro número: "))
sinal = input("insira a operação matemática + ou -: ")
num_2 = int(input("Insira o segundo número: "))

#Comandos if e else
if sinal == "+":
  print(f"O resultado da adição é: {num_1 + num_2}")
elif sinal =="-":
  print(f"O resultado da subtração é: {num_1 - num_2}")
else:
  print(f"Desculpe, só sei calcular adição e subtração")
```

**Calculadora com casas decimais**

```python
#Nome da Calculadora
print("Calculadora de adição ou subtração de dois produtos, com casas decimais")

#Variáveis (utilizando o float para números com casas decimais)
num_1 = float(input("Insira o primeiro número, com casas decimais: "))
sinal = input("insira a operação matemática + ou -: ")
num_2 = float(input("Insira o segundo número, com casas decimais: "))

#Comandos if e else
if sinal == "+":
  print(f"O resultado da adição é: {num_1 + num_2}")
elif sinal =="-":
  print(f"O resultado da subtração é: {num_1 - num_2:.2f}")
else:
  print(f"Desculpe, só sei calcular adição e subtração")
```

**Calculadora de IMC**

```python
#Nome da Calculadora

#Criando as variaveis com input
nome = input("Digite seu nome:" )
idade = input("Informe sua idade: ")
peso = float(input("Qual seu peso (em KG)? "))
altura = float(input("Qual sua Altura (em metros e usando ponto)? "))

imc = float(peso / (altura*altura))

#Mensagem 1
#print(f'Olá, {nome}!')
#print(f'seu IMC é: {imc}')

if imc < 16:
	print(f"{nome}, seu estado é de Magreza grave. Procure um médico ou nutricionista.")
elif imc < 17:
	print(f"{nome}, seu estado é de Magreza moderada. Precisa rever sua alimentação.")
elif imc < 18.5:
	print(f"{nome}, seu estado é de Magreza leve. Precisa rever sua alimentação.")
elif imc < 25:
	print(f"Você está saudável {nome}. Parabéns!")
elif imc < 30:
	print(f"{nome}, seu estado é de Sobrepeso. Precisa rever sua alimentação.")
elif imc < 35:
	print(f"{nome}, seu estado é de Obesidade Grau I. Procure um médico ou nutricionista.")
elif imc < 40:
	print(f"{nome}, seu estado é de Obesidade Grau II (severa). Procure um médico ou nutricionista.")
else:
	print(f"{nome}, seu estado é de Obesidade Grau III (mórbida). Procure um médico ou nutricionista.")
```
