```python
# Exemplo 03 - Retorna se um valor é par ou não.

def epar(x):
  return(x % 2 == 0)

print(epar(2))
print(epar(3))
print(epar(10))
```

```python
# Exemplo 04 - Reutiliza a função epar em outra função.

def epar(x):
  return(x % 2 == 0)

def par_ou_impar(x):
  if epar(x):
    return "par"
  else:
    return "ímpar"

print(par_ou_impar(4))
print(par_ou_impar(5))
```

# 1) Faça um programa com uma função que retorne o maior de dois números.

```python
# 1 - Faça um programa com uma função que retorne o maior de dois números.

def numero_maior(a, b):
    if a > b:
        return a
    else:
        return b

numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))

resultado = numero_maior(numero1, numero2)
print(f"\nO maior número é: {resultado}")
```

# 2) Faça um programa com uma função que receba dois números e retorne True se o primeiro número for múltiplo do segundo.


```python
# 2 - Faça um programa com uma função que receba dois números e retorne True se o primeiro número for múltiplo do segundo.

def multiplo(numero1, numero2):
    if numero1 % numero2 == 0:
        return True
    else:
        return False

numero1 = int(input("Digite o primeiro número: "))
numero2 = int(input("Digite o segundo número: "))

if multiplo(numero1, numero2):
    print(f"\n{numero1} é múltiplo de {numero2}.")
else:
    print(f"\n{numero1} não é múltiplo de {numero2}.")
```

# 3) Faça um programa com uma função que receba o lado (l) de um quadrado e retorne a sua área (A = lado2).

```python
# 3 - Faça um programa com uma função que receba o lado (l) de um quadrado e retorne a sua área (A = lado²).

def calcular_area(lado):
    area = lado ** 2
    return area

lado = float(input("Digite o valor do lado do quadrado: "))
area_total = calcular_area(lado)
print(f"A área do quadrado é: {area_total:.2f}")
```

# 4) Faça um programa com uma função que receba a base e a altura de um triângulo e retorne a sua área (A = (base x altura) / 2).


```python
# 4 - Faça um programa com uma função que receba a base e a altura de um triângulo e retorne a sua área (A = (base x altura) / 2).

def calcular_triangulo(base, altura):
    area = (base * altura) / 2
    return area

base_triangulo = float(input("Digite o valor da base do triângulo: "))
altura_triangulo = float(input("Digite o valor da altura do triângulo: "))

area_resultante = calcular_triangulo(base_triangulo, altura_triangulo)
print(f"A área do triângulo é: {area_resultante}")
```

# 5) Faça um programa com uma função que receba o raio (R) de um círculo e retorne a sua área (A = PI * R²).


```python
# 5 - Faça um programa com uma função que receba o raio (R) de um círculo e retorne a sua área (A = PI * R²).

import math

def calcular_area_circulo(raio):
    area = math.pi * raio ** 2
    return area

raio_circulo = float(input("Digite o valor do raio do círculo: "))
area_resultante = calcular_area_circulo(raio_circulo)
print(f"A área do círculo é: {area_resultante:.2f}")
```

# 7) Faça um programa com uma função chamada somaImposto. A função possui dois parâmetros formais: taxaImposto, que é a quantia de imposto sobre vendas expressa em porcentagem, e custo, que é o custo de um item antes do imposto. A função "altera" o valor de custo para incluir o imposto sobre vendas.


```python
# 7 - Faça um programa com uma função chamada somaImposto. A função possui dois parâmetros formais: taxaImposto, que é a quantia de imposto sobre vendas
# expressa em porcentagem, e custo, que é o custo de um item antes do imposto. A função "altera" o valor de custo para incluir o imposto sobre vendas.

def somaImposto(taxaImposto, custo):
    custo_com_imposto = custo * (1 + taxaImposto / 100)
    return custo_com_imposto

taxa = float(input("Digite a taxa de imposto (%): "))
valor = float(input("Digite o custo do item: R$ "))
valor_com_imposto = somaImposto(taxa, valor)
print(f"\nCusto com imposto: R$ {valor_com_imposto:.2f}")
```

# 8) Escreva um programa, com uma função para ler a idade de uma pessoa e informar a sua classe eleitoral, conforme:
  * • Não-eleitor (abaixo de 16 anos).
  * • Eleitor obrigatório (entre 18 e 65 anos).
  * • Eleitor facultativo (entre 16 e 18 ou maior de 65 anos).


```python
# 8 - Escreva um programa, com uma função para ler a idade de uma pessoa e informar a sua classe eleitoral, conforme:
      # Não-eleitor (abaixo de 16 anos).
      # Eleitor obrigatório (entre 18 e 65 anos).
      # Eleitor facultativo (entre 16 e 18 ou maior de 65 anos).

def classe_eleitoral(idade):
    if idade < 16:
        return "Não-eleitor"
    elif 16 <= idade < 18 or idade >= 65:
        return "Eleitor facultativo"
    else:
        return "Eleitor obrigatório"

idade = int(input("Digite a idade: "))
classe = classe_eleitoral(idade)
print("\nClasse eleitoral:", classe)
```

# 9) Escreva um programa, com uma função que imprima o conceito de um aluno, dada a sua nota. Supor, apenas, notas inteiras. O critério para conceitos é o seguinte:
  * Inferiores à 3  | Conceito E
  * De 3 à 5        | Conceito D
  * Notas 6 e 7     | Conceito C
  * Notas 8 e 9     | Conceito B
  * Nota 10         | Conceito A


```python
# 9 - Escreva um programa, com uma função que imprima o conceito de um aluno, dada a sua nota. Supor, apenas, notas inteiras. O critério para conceitos é o seguinte:
      # Inferiores à 3  | Conceito E
      # De 3 à 5        | Conceito D
      # Notas 6 e 7     | Conceito C
      # Notas 8 e 9     | Conceito B
      # Nota 10         | Conceito A

def conceito_aluno(nota):
    if nota < 3:
        return "Conceito E"
    elif 3 <= nota <= 5:
        return "Conceito D"
    elif 6 <= nota <= 7:
        return "Conceito C"
    elif 8 <= nota <= 9:
        return "Conceito B"
    else:
        return "Conceito A"

nota_aluno = int(input("Digite a nota do aluno: "))
conceito = conceito_aluno(nota_aluno)
print("\nConceito do aluno:", conceito)
```

# 10) Escreva um programa, com uma função que leia 2 valores (A e B), calcule e mostre a soma dos números ímpares entre eles (inclusive A e B). Nesse caso, considere que só serão informados valores inteiros positivos e que A é menor do que B.

```python
# 10 - Escreva um programa, com uma função que leia 2 valores (A e B), calcule e mostre a soma dos números ímpares entre eles (inclusive A e B).
# Nesse caso, considere que só serão informados valores inteiros positivos e que A é menor do que B.

def soma_impares_entre_numeros(a, b):
    soma = 0
    for num in range(a, b + 1):
        if num % 2 != 0:
            soma += num
    return soma

valor_a = int(input("Digite o valor de A: "))
valor_b = int(input("Digite o valor de B: "))
resultado_soma = soma_impares_entre_numeros(valor_a, valor_b)
print(f"\nSoma dos ímpares entre A e B: {resultado_soma}")
```
