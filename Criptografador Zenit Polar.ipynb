```python
from time import sleep

print("Bem vindo ao encriptador ZP by reises.py\n")

senha_correta = 'zenitpolar' #futuramente irei inserir uma página de cadastro e um banco de dados de senha...
senha = input("Digite a senha da criptografia: ")

if senha == senha_correta:
  sleep(0.5)
  print(f"Checando sua senha...\n")
  sleep(1)
  print(f"Entrada autorizada!")
  mensagem = input("\nDigite sua mensagem: ")
  ZPCRIPT = {
    'Z': 'P', 'P': 'Z', 'E': 'O', 'O': 'E', 'N': 'L', 'L': 'N', 'I': 'A', 'A': 'I', 'T': 'R', 'R': 'T',
    'z': 'p', 'p': 'z', 'e': 'o', 'o': 'e', 'n': 'l', 'l': 'n', 'i': 'a', 'a': 'i', 't': 'r', 'r': 't',
  }
  resultado = ""
  for i in range(len(mensagem)):
    if mensagem[i] in ZPCRIPT:
      resultado += ZPCRIPT[mensagem[i]]
    else:
      resultado += mensagem[i]

  print(f"\nAguarde enquanto traduzimos sua mensagem...\n")
  sleep(1)
  print(f"\nMensagem traduzida:\n {resultado}")
else:
  sleep(0.5)
  print(f"Checando sua senha...")
  sleep(3)
  print("\nSenha incorreta.\nVocê não possui permissão para acessar.")
```
