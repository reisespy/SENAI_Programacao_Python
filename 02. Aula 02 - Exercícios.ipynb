4. Criar o fluxograma para calcular a média de um aluno e mostrar
o status do mesmo. Média >= 6.0; Status ← “Aprovado”; Média <
6.0; Status ← “Reprovado”;

```python
#4. Criar code para calcular a média de um aluno e mostrar o status do mesmo.
#Média >= 6.0; Status ← “Aprovado”; Média < 6.0; Status ← “Reprovado”;

    #Apresentando a calculadora de médias
print("Calculadora de média anual do aluno")

print(" ")

    #Criando as variáveis
nome = input("Insira o nome do aluno: ")
disciplina = input("Insira a disciplina: ")
print(" ")
nota1 = float(input("Insira a nota do 1º bimestre: "))
nota2 = float(input("Insira a nota do 2º bimestre: "))
nota3 = float(input("Insira a nota do 3º bimestre: "))
nota4 = float(input("Insira a nota do 4º bimestre: "))
media = ((nota1 + nota2 + nota3 + nota4) / 4)

print(" ")
    #Usando comandos if e else
if media>=8.0:
  print(f"Aluno {nome} está ***Aprovado** com excelência na disciplina de {disciplina} :)")
elif(media>=6.0 and media<7.0):
  print(f"Aluno {nome} está **Aprovado** na disciplina de {disciplina}, mas foi por pouco hein...")
else:
  print(f"Aluno {nome} está **Reprovado** vergonhosamente na disciplina de {disciplina} :(")

print(" ")
    #Informando a média
print(f"A média final do aluno {nome} é: {media:.2f}")
```
