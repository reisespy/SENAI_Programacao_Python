```python
import csv

def consumo_por_utilizacao(potencia_chuveiro, tempo_utilizacao_minutos):
    consumo_kwh = (potencia_chuveiro * tempo_utilizacao_minutos) / 60000
    return consumo_kwh

def consumo_diario(consumo_por_banho, numero_banhos_diarios):
    return consumo_por_banho * numero_banhos_diarios

def consumo_semanal(consumo_diario_kwh):
    return consumo_diario_kwh * 7

def consumo_mensal(consumo_diario_kwh):
    return consumo_diario_kwh * 30

def main():
  # Será substituído pelo log.txt ou csv do registro realizado a partir
  # do sistema de cadastro do produto e sistema de cronometragem
    potencia_chuveiro = int(input("Digite a potência do chuveiro (Watts): "))
    numero_banhos_diarios = int(input("Digite o número de banhos diários: "))

    consumo_total_diario = 0
    registro_utilizacao = []

    for i in range(numero_banhos_diarios):
        nome_usuario = input(f"Digite o nome do usuário para o banho {i + 1}: ")
        tempo_utilizacao_minutos = int(input(f"Digite o tempo de utilização do banho {i + 1} (minutos): "))
        consumo_banho = consumo_por_utilizacao(potencia_chuveiro, tempo_utilizacao_minutos)
        consumo_total_diario += consumo_banho
        registro_utilizacao.append({"nome": nome_usuario, "consumo": consumo_banho})
        print(f"Consumo do banho {i + 1}: {consumo_banho:.2f} kWh")

    consumo_total_semanal = consumo_semanal(consumo_total_diario)
    consumo_total_mensal = consumo_mensal(consumo_total_diario)

    print("\nResultados:")
    print(f"Consumo diário: {consumo_total_diario:.2f} kWh")
    print("")
    print(f"Previsão de Consumo semanal: {consumo_total_semanal:.2f} kWh")
    print(f"Previsão de Consumo mensal: {consumo_total_mensal:.2f} kWh")

    print("\nRegistro de utilização:")
    for i, registro in enumerate(registro_utilizacao, 1):
        print(f"Banho {i}: Usuário: {registro['nome']}, Consumo: {registro['consumo']:.2f} kWh")

    # Salvar registros em um arquivo CSV
    with open("registro_diario.csv", mode="w", newline="") as file:
        fieldnames = ["banho", "nome", "consumo_kwh"]
        writer = csv.DictWriter(file, fieldnames=fieldnames)

        writer.writeheader()
        for i, registro in enumerate(registro_utilizacao, 1):
            writer.writerow({"banho": i, "nome": registro["nome"], "consumo_kwh": registro["consumo"]})

if __name__ == "__main__":
    main()
```
