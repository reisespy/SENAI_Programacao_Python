# 2. Faça um programa em Python (utilize a estrutura for) que leia 10 valores inteiros e:
* Encontre e mostre o maior valor
* Encontre e mostre o menor valor
* Calcule e mostre a média dos números lidos

```python
# Inicializando variáveis para armazenar o maior, menor e soma dos valores lidos
maior_valor = float("-inf")
menor_valor = float("inf")
soma_valores = 0

# Utilizando o laço "for" para ler os 10 valores inteiros
for i in range(1, 11):
    try:
        valor = int(input(f"Digite o {i}º valor inteiro: "))
        # Atualizando o maior e menor valor, se necessário
        if valor > maior_valor:
            maior_valor = valor
        if valor < menor_valor:
            menor_valor = valor
        # Somando os valores para calcular a média
        soma_valores += valor
    except ValueError:
        print("Valor inválido. Digite apenas números inteiros.")

# Calculando a média
media = soma_valores / 10

# Mostrando os resultados
print(f"\nMaior valor: {maior_valor}")
print(f"Menor valor: {menor_valor}")
print(f"Média dos números lidos: {media}")
```

# 3. Faça um programa em linguagem Python que recebe a temperatura de z cliente e imprima a mensagem de se a temperatura esta normal (menor que 37,2 C) ou está em estado febril (37,3 C a 38 C ) ou com febre (38 C a 39 C) e com febre alta(acima 39 C). No final mostre a quantidade de pessoas analisadas e a média de temperatura.

```python
def classificar_temperatura(temperatura):
    if temperatura <= 35:
        return "Estado de hipotermia"
    elif 35.1 <= temperatura <= 36:
        return "Temperatura muito baixa"
    elif 36.1 <= temperatura <= 37.2:
        return "Temperatura normal"
    elif 37.3 <= temperatura < 38:
        return "Estado febril"
    elif 38 < temperatura < 39:
        return "Com febre"
    elif 39.1 <= temperatura < 40:
        return "Com febre alta"
    else:
        return "TA PEGANDO FOGO BIXO!!!"

def relatorio():
    num_clientes = int(input("Quantas pessoas deseja analisar? "))
    soma_temperaturas = 0

    for i in range(1, num_clientes + 1):
        try:
            temperatura_cliente = float(input(f"\nDigite a temperatura do cliente {i}: "))
            soma_temperaturas += temperatura_cliente
            classificacao = classificar_temperatura(temperatura_cliente)
            print(f"Cliente {i}: {classificacao}\n")
        except ValueError:
            print("Valor inválido. Digite apenas números.")

    media_temperaturas = soma_temperaturas / num_clientes
    print(f"Quantidade de pessoas analisadas: {num_clientes}")
    print(f"Média de temperatura: {media_temperaturas:.2f} ºC")

if __name__ == "__main__":
  relatorio()
```

# 4. Ler do teclado a idade e o sexo de 10 pessoas, calcule e imprima:
* idade média das mulheres
* idade média dos homens
* idade média do grupo

```python
def mediaidades():
    num_pessoas = 10
    soma_idade_mulheres = 0
    soma_idade_homens = 0
    soma_idade_grupo = 0
    num_mulheres = 0
    num_homens = 0

    for i in range(1, num_pessoas + 1):
        try:
            idade = int(input(f"Digite a idade da pessoa {i}: "))
            sexo = input(f"Digite o sexo da pessoa {i} (M para masculino, F para feminino): ").upper()

            if sexo == 'F':
                soma_idade_mulheres += idade
                num_mulheres += 1
            elif sexo == 'M':
                soma_idade_homens += idade
                num_homens += 1
            elif sexo != 'M' or 'F':
                print("Digite um sexo válido.")
                continue

            soma_idade_grupo += idade
        except ValueError:
            print("Valor inválido. Digite apenas números para a idade.")

    media_idade_mulheres = soma_idade_mulheres / num_mulheres if num_mulheres > 0 else 0
    media_idade_homens = soma_idade_homens / num_homens if num_homens > 0 else 0
    media_idade_grupo = soma_idade_grupo / num_pessoas

    print(f"\nIdade média das mulheres: {media_idade_mulheres:.2f}")
    print(f"Idade média dos homens: {media_idade_homens:.2f}")
    print(f"Idade média do grupo: {media_idade_grupo:.2f}")

if __name__ == "__main__":
  mediaidades()
```

# 5. Escreva um programa que mostre todos os números entre 5 e 100 que são divisíveis por 7, mas não são múltiplos de 5. Os números obtidos devem ser impressos em sequência.

```python
numeros_divisiveis = []

for num in range(5, 100):
  if num % 7 == 0 and num % 5 != 0:
    numeros_divisiveis.append(num)

print("Números divisíveis por 7 e não múltiplos de 5:")
print(numeros_divisiveis)
```

# 7. Escreva um programa em Python para encontrar o fatorial de qualquer número.

```python
def calcular_fatorial(num):
    if num == 0 or num == 1:
        return 1
    else:
        return num * calcular_fatorial(num - 1)

def fatorial():
    try:
        numero = int(input("Digite um número inteiro para calcular o fatorial: "))
        if numero < 0:
            print("O fatorial é válido apenas para números inteiros não negativos.")
        else:
            resultado = calcular_fatorial(numero)
            print(f"O fatorial de {numero} é: {resultado}")
    except ValueError:
        print("Entrada inválida. Digite um número inteiro.")

if __name__ == "__main__":
  fatorial()
```

# 8. Faça um programa que leia um nome de usuário e a sua senha e não aceite a senha igual ao nome do usuário, mostrando uma mensagem de erro e voltando a pedir as informações.

```python
from time import sleep

def obter_credenciais():
    while True:
        usuario = input("Digite o nome de usuário: ")
        senha = input("Digite a senha: ")

        if senha != usuario:
            return usuario, senha
        else:
            print("A senha não pode ser igual ao nome de usuário. Tente novamente.\n")

def registrook():
    print("Registro de Usuário\n")
    usuario, senha = obter_credenciais()
    print("Aguarde enquanto registramos em nosso banco de dados...")
    sleep(2)
    print("Tudo pronto!")
    sleep(1)
    print("\nUsuário registrado com sucesso!")
    print(f"Nome de usuário: {usuario}")
    print(f"Senha: {senha}")

if __name__ == "__main__":
  registrook()
```
